```python
result1 = []
result2 = []
```

```python
n1 = 0
n2 = 0
for i in range(len(lines)):
    if lines[i].endswith("f5f7\n"):
        n1 += 1
        result1.append(n1/(i+1))
    if lines[i].endswith("e6f7\n"):
        n2 += 1
        result2.append(n2/(i+1))        
```

```python
plt.plot(result1)
plt.plot(result2)
```
